```python
from PIL import Image
import pytesseract 
```

```python
img_path = 'cv-sample-1.png'
img = Image.open(img_path)
img
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
text = pytesseract.image_to_string(img)
print(text)
```

```python
key = ['PENGALAMAN', 'EXPERIENCE', 'SKILLS', 'KEAHLIAN', 'PENDIDIKAN', 'EDUCATION', 'PENGHARGAAN', 'ACHIEVEMENT', 'COURSE PROJECTS', 'WORK EXPERIENCE', ]
key
```

```python
text_list = text.split('\n')
text_list
```

```python
splited = text_list.copy()
section = ['PROFILE']

for i in range(len(text_list)):
    for k in key:
        if text_list[i] == k:
            splited[i] = 'SECTION'
            section.append(text_list[i])
```

```python
result = ('\n'.join(splited)).split('SECTION')
```

```python
dic = {}

for i in range(len(result)):
    dic[section[i]] = result[i]
```

```python
print(dic)
```
